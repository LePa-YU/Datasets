```
import random
def random_six(low_limit, hi_limit):
  six = []
  while len(six) < 7:
    rand = random.randint(low_limit, hi_limit)
    if rand in six:
      continue
    else:
      six.append(rand)
  return six

print(random_six(1,10))
```

```
def grade_points(letter_grade):
  grades = [('A+',9),('A',8),('B+',7),('B',6),('C+',5),('C',4),('D+',3),('D',2),('E',1),('F',0)]

  for i in grades:
    grade,number = i
    if grade is letter_grade:
      return number

  return -1
print(grade_points('A'))
```

```
def sum_sq_digits(n):
  n = str(n)
  squared = list(n)
  num = 0
  for i in squared:
    num = num + int(i)*int(i)
  
  return num

print(sum_sq_digits(345))
```

```
import random
def roll():
  return random.randint(1,6)

def count(value):
  sum = 0
  for i in range(0, value):
    sum = sum+roll()
  return sum

def main():
  while True:
    try:
      value = int(input("Give input"))
    except ValueError:
      print("ValueError")
      value = 0
    if value == 0:
      break
    print(count(value))

main()
```

```
from datetime import datetime, timedelta
def date(value):
  now = datetime.today()
  return now + timedelta(days = value)

def weekday(date):
  days = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
  return days[date.weekday()]

def main():
  while True:
    try:
      value = int(input("Give input"))
    except ValueError:
      print("ValueError")
      value = 0
    if value == 0:
      break
    print(weekday(date(value)))

main()
```
